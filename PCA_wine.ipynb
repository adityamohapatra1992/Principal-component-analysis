```python
# split data to X and Y Variables

X = df.iloc[:,1:]
X.head()
```

```python
Y = df.iloc[:,0]
Y.head()
```

```python
# Data Standardization

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
X_std = scaler.transform(X)
X_std
```

```python
# We will now Perform Dimensionality Reduction with PCA.

from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca.fit_transform(X_std)
```

```python
PCA_TF = pca.transform(X_std)
```

```python
var = pca.explained_variance_
print(var)
```

```python
pca.components_[0]
```

```python
pca.n_features_
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10,8))
plt.plot(pca.explained_variance_ratio_.cumsum(),marker = 'o',linestyle = '--')
plt.title('Explained variance by component')
plt.xlabel('Number of components')
plt.ylabel('Cumulative Explained variance')
```

```python
# We will choose 2 to 3 components 

pca = PCA(n_components=3)
```

```python
PR_comp = pca.fit_transform(X_std)
PR_comp
```

```python
Principal_components = pd.DataFrame(data=PR_comp,columns=['Principal Component 1','Principal Component 2','Principal Component 3'])
Principal_components.head()
```

```python
# We will now perform Kmeans clustering

from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(1,20):
    Kmeans_pca = KMeans(n_clusters = i,init='k-means++',random_state=42)
    Kmeans_pca.fit(PR_comp)
    wcss.append(Kmeans_pca.inertia_)
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,20),wcss,marker='o',linestyle='--')
plt.xlabel('Number of cluster')
plt.ylabel('wcss')
plt.title('Kmeans clustering with PCA')
```

```python
# Now that we know there are 3 clusters.

Kmeans_new = KMeans(n_clusters=3,init='k-means++',random_state=42)
```

```python
a = Kmeans_new.fit_predict(PR_comp)
a
```

```python
# Visualize the cluster using scatterplot
plt.figure(figsize=(10,8))
sns.scatterplot(PR_comp[a==0,0],PR_comp[a==0,1],label='cluster 1')
sns.scatterplot(PR_comp[a==1,0],PR_comp[a==1,1],label='cluster 2')
sns.scatterplot(PR_comp[a==2,0],PR_comp[a==2,1],label='cluster 3')
# create the centroids.
sns.scatterplot(Kmeans_new.cluster_centers_[:,0],Kmeans_new.cluster_centers_[:,1],label='Centroids',color='black')
plt.title('Clusters by PCA Components')
```

```python
import scipy.cluster.hierarchy as sch  
from scipy.cluster.hierarchy import linkage, dendrogram 
```

```python
z = linkage(PR_comp, method='complete', metric='euclidean')
```

```python
plt.figure(figsize=(25,5))
plt.title('Hierarchial Dendogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,leaf_font_size=8,leaf_rotation=0)
plt.show()
```

```python
z = linkage(PR_comp, method='ward', metric='euclidean',optimal_ordering=False)
plt.figure(figsize=(25,10))
plt.title('Hierarchial of the PCA Components')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,leaf_font_size=8,leaf_rotation=0)
plt.show()
```
